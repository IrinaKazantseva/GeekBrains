## Домашнее задание 3

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.


### 1. Построим матрицу корреляции

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.5] = 0
corr_matrix
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

Видим, что наиболее коррелируемые признаки - это:
1. sqft_above и sqft_living. 
2. sqft_living и price.
3. sqft_living и bathrooms.
4. sqft_living и grade.


### 2. Посмотрим на зависимость цены от площади

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

Присутствует прямая зависимость

### 3. Посмотрим на зависимость цены от оценки

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

В целом связь прослеживается, но присутствует большое количество выбросов

### 4. Изучим зависимость цены от года постройки

```python
plt.figure(figsize=(10, 100))

sns.boxplot(x=df['price'], y=df['yr_built'].astype('str'), whis=1.5)

plt.xlabel('price')
plt.ylabel('yr_built')
plt.title('Distribution of price by yr_built');
```

Четкой связи не прослеживается, много выбросов, медианные значения в пределах одной величины

### 5. Посмотрим на зависимость цены от количества спален

```python
plt.figure(figsize=(10, 12))

sns.boxplot(x=df['price'], y=df['bedrooms'].astype('str'), whis=1.5)

plt.xlabel('price')
plt.ylabel('bedrooms')
plt.title('Distribution of price by bedrooms');
```

Выбросов значительное количество, но в целом диапазон цен выше у домов с 8 спальнями
